```python
import pandas as pd
from collections import Counter
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
df = pd.read_json('all.json')
```

```python
texts = list(df['text'])
```

```python
c = Counter()
for text in texts:
    for sent in sent_tokenize(text):
        for word in word_tokenize(sent):
            c[word.lower()] += 1
```
